## Check for missing values in all columns and replace them with the appropriate metric

```python
df["CustomerID"].fillna(df["CustomerID"].mode()[0], inplace=True)
```

We used mode because we want to maintain unifoirmity of the data so the customer ID which is most frequent we relpace all the null values with the mode of column "CustomerID".

## Remove duplicate rows

```python
df = df.drop_duplicates()
df
```

## Remove rows which have negative values in Quantity column

```python
df.drop(df[df['Quantity'] < 0].index, inplace=True)
df
```

## Add the columns - Month, Day and Hour for the invoice

```python
df['Month'] = pd.DatetimeIndex(df['InvoiceDate']).month
df['Day'] = pd.DatetimeIndex(df['InvoiceDate']).day
df['Hours'] = pd.DatetimeIndex(df['InvoiceDate']).hour
df
```
